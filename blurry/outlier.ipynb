```python
x = (torch.rand(11) - 0.5) * 4
y = x**2 + torch.randn(11) * 0.3
y[5] = 4

plt.scatter(x, y)
plt.xlim([-3, 3])
plt.ylim([-0.5, 4.5])
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(x.view(-1, 1))

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
print(lin_reg.intercept_ , lin_reg.coef_)

test_x = torch.linspace(-2.0, 2.0, 101)
pred_y = lin_reg.predict(poly_features.transform(test_x.view(-1, 1)))

plt.scatter(x, y)
plt.plot(test_x, pred_y, color='r')
plt.xlim([-3, 3])
plt.ylim([-0.5, 4.5])
```
