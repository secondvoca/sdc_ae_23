```python
import cv2
import matplotlib.pyplot as plt

path = '/home/secondvoca/sdc_ae_23/data/ip_data/Lenna.png'
img = cv2.imread(path, cv2.IMREAD_COLOR)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

# histogram equalization

```python
import numpy as np

hist, _ = np.histogram(img_gray.reshape(-1), bins=256)

# accumulative sum
eq_vals = (hist.cumsum() / hist.sum() * 255).astype(int)

img_new = np.zeros_like(img_gray)
for r in range(img_gray.shape[0]):
    for c in range(img_gray.shape[1]):
        img_new[r, c] = eq_vals[img_gray[r, c]]

_, (ax_1, ax_2) = plt.subplots(1, 2)
ax_1.imshow(img_gray, cmap='gray')
ax_1.axis('off')
ax_2.imshow(img_new, cmap='gray')
ax_2.axis('off')
plt.tight_layout()
plt.show()
```
